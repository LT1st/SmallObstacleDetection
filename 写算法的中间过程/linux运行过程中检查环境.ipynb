# GPU使用情况

禁用nouveau模块

%%!
lsmod |grep -i nouveau

## 1:nvidia-smi报错

详细描述：报错信息为NVIDIA-SMI has failed because it couldn't communicate with the NVIDIA driver. Make sure that the latest NVIDIA driver is installed and running.

原因：系统自动更新了内核版本，重启后使用了新的内核，但显卡驱动还是关联旧版本的内核，所以找不到

解决方案：重新关联新的内核和显卡驱动

%%!
cd /usr/src && ls -d nvidia-* #获得驱动版本，如nvidia-465.19.01
apt install dkms #安装dkms
dkms install -m nvidia -v 465.19.01 #重新关联

## 2: 启后找不到”/dev/nvidia-uvm”
详细描述：报错信息为Error response from daemon: error gathering device information while adding custom device "/dev/nvidia-uvm: no such file or directory

原因：系统重启后，nvidia部分模块没有挂载上

解决方案：重新挂载所有模块

%%!
cd /usr/local/cuda/samples/1_Utilities/deviceQuery && make
./deviceQuery
